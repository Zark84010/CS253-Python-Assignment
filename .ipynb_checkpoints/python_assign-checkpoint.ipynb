### Data pre-processing
The first step in every machine learning algorithm is to process the raw data in some meaningful representations. We will be using the [Bag-of-Words](https://towardsdatascience.com/a-simple-explanation-of-the-bag-of-words-model-b88fc4f4971) representation to process the text. It comprises of following steps:

- Process emails line-by-line to extract all the words.
- Replace extracted words by their stem (root) word. This is known as stemming and lematization.
- Remove stop words like and, or, is, am, and so on.
- Assign a unique index to each word. This forms the vocabulary.
- Represent each email as a binary vector of length equal to the size of the vocabulary such that the $i^{th}$ element of the vector is 1 iff the $i^th$ word is present in the email.

Here we provide you with the function signature along with the expected functionality. You are expected to complete them accordingly. 

```python
import nltk
nltk.download('stopwords')
```

```python
import numpy as np
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords

# takes an email as an argument
# read email line-by-line and extract all the words
# return list of extracted words
def read_email(email):
    email = email.replace('_', '')
    email = email.strip()
    words = email.split(' ')
    words = [word.lower() for word in words]
    words = [word for word in words if word != ' ' and word != '']
    return words
  
# takes a list of words as an argument
# replace each word by their stem word
# return list of stem words
def stemming(words):
    ps = PorterStemmer()
    stem_words = [ps.stem(word) for word in words]
    return stem_words

# takes a list of stem-words as an argument
# remove stop words
# return list of stem words after removing stop words
stopWords = set(stopwords.words('english'))
def remove_stop_words(stem_words):
    stem_no_stop_words = [stem_word for stem_word in stem_words if stem_word not in stopWords]
    return stem_no_stop_words

# takes a list of stem-words as an argument
# add new words to the vocabulary and assign a unique index to them
# returns new vocabulary
vocab = dict()
def build_vocabulary(stem_words):
    idx = len(vocab)
    for stem_word in stem_words:
        if stem_word not in vocab.keys():
            vocab[stem_word] = idx
            idx += 1

# takes a list of stem-words and vocabulary as an argument
# returns bow representation
def get_bow(swords):
    swords_set = set(swords)
    email_bow = [1 if word in swords_set else 0 for word in vocab.keys()]
    return email_bow

def text_to_bow(email):
    swords = remove_stop_words(stemming(read_email(email)))
    build_vocabulary(swords)
    return get_bow(swords)

# read the entire dataset
# convert emails to bow and maintain their labels
# call function text_to_bow()
def read_data():
    bows = []
    with open("spam_or_not_spam.csv", "r") as dataset:
        cnt = 0
        for idx, line in enumerate(dataset):
            if idx > 0:
                email, label = line.strip().split(',')
                bow = text_to_bow(email)
        dataset.seek(0)
        for idx, line in enumerate(dataset):
            if idx > 0:
                email, label = line.strip().split(',')
                label = int(label)
                bow = text_to_bow(email)
                bow.append(label)
                bows.append(bow)
    data = np.array(bows, dtype=bool)
    return data
```

```python
data = read_data()
```

### Data Visualization
Let's understand the data distribution
- Visualize the frequency of word-occurence in all the emails(spam + non-spam)
- Visualize the freuency of word-occurence for spam and non-spam emails separately

```python
import matplotlib.pyplot as plt

# visuallze data distribution

vocab_arr = [''] * len(vocab)
for word in vocab:
    idx = vocab[word]
    vocab_arr[idx] = word

def plot_data(data, num_words, heading):
    print(heading)
    print(f"Total: {data.shape[0]}")
    print(f"Top {num_words} frequent words")
    sums = data.sum(axis=0)
    indices = np.argsort(sums)[::-1][:num_words]
    words = [vocab_arr[idx] for idx in indices]
    nums = [sums[idx] for idx in indices]
    plt.figure(figsize=(20, 10))
    bar_plot = plt.bar(range(num_words), nums, align='center')
    plt.xlabel("Words")
    plt.ylabel("Frequency")
    plt.xticks(range(num_words), words)
    y_off = 0.012 * nums[0]
    for bar in bar_plot:
        yval = bar.get_height()
        plt.text(bar.get_x(), yval+y_off, yval)
    plt.show()

def data_vis(data):
    data_all = data[:, :-1]
    data_good = data[data[:, -1] == 0][:, :-1]
    data_spam = data[data[:, -1] == 1][:, :-1]
    plot_data(data_all, 20, 'All Emails')
    plot_data(data_good, 20, 'Regular Emails')
    plot_data(data_spam, 20, 'Spam Emails')

data_vis(data)
```

### Learn a Classifier
Split the dataset randomly in the ratio 80:20 as the training and test dataset. Use only training dataset to learn the classifier. No test data should be used during training. Test data will only be used during evaluation.

Now let us try to use ML algorithms to classify emails as spam or non-spam. You are supposed to implement [SVM](https://scikit-learn.org/stable/modules/svm.html) and [K-Nearest Neighbour](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html) algorithm available in scikit-learn using the same training dataset for both.

```python
num_top = 1000

indices = np.append(np.argsort(data[:, :-1].sum(axis=0))[::-1][:num_top], -1)
data = data[:, indices]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm

# split dataset
def split(data):
    num_all = len(data)
    num_train = round(0.8 * num_all)
    num_test = num_all - num_train
    np.random.shuffle(data)
    train_data, test_data = data[:num_train, :], data[num_train:, :]
    return train_data, test_data

# learn a SVM model
# use the model to make prediction
# return the model predictions on train and test dataset

def get_X_y(data):
    return data[:, :-1], data[:, -1]

def classifier_preds(train_data, test_data, classifier):
    X_train, y_train = get_X_y(train_data)
    X_test,  y_test  = get_X_y(test_data)
    
    classifier.fit(X_train, y_train)
    
    predict_labels = classifier.predict(X_train), classifier.predict(X_test)
    
    return predict_labels

def svm_classifier(train_data, test_data):
    classifier = svm.SVC()
    return classifier_preds(train_data, test_data, classifier)

# implement k-NN algorithm
# use the model to make prediction
# return the model predictions on train and test dataset
def knn_classifier(train_data, test_data):
    classifier = KNeighborsClassifier()
    return classifier_preds(train_data, test_data, classifier)

train_data, test_data = split(data)
svm_train_predictions, svm_test_predictions = svm_classifier(train_data, test_data)
knn_train_predictions, knn_test_predictions = knn_classifier(train_data, test_data)
```

### Model Evaluation
Compare the SVM and k-NN model using metrics
- Accuracy
- [AUC score](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.auc.html)


```python
from sklearn import metrics

# compute accuracy 
def compute_accuracy(true_labels, predicted_labels):
    acc = metrics.accuracy_score(true_labels, predicted_labels)
    return acc

# compute AUC score 
def compute_auc(true_labels, predicted_labels):
    auc = metrics.roc_auc_score(true_labels, predicted_labels)
    return auc

# write code to print train and test accuracy and AUC score of SVM and k-NN classifier
train_true = train_data[:, -1]
test_true = test_data[:, -1]

print('SVM Metrics')
svm_acc = compute_accuracy(test_true, svm_test_predictions)
svm_auc = compute_auc(test_true, svm_test_predictions)
svm_train_acc = compute_accuracy(train_true, svm_train_predictions)
svm_train_auc = compute_auc(train_true, svm_train_predictions)
print('')
print(f'SVM Accuracy: {svm_acc * 100} %')
print(f'SVM AUC: {svm_auc * 100} %')
print(f'SVM Train Accuracy: {svm_train_acc * 100} %')
print(f'SVM Train AUC: {svm_train_auc * 100} %')

print('')

print('KNN Metrics')
knn_acc = compute_accuracy(test_true, knn_test_predictions)
knn_auc = compute_auc(test_true, knn_test_predictions)
knn_train_acc = compute_accuracy(train_true, knn_train_predictions)
knn_train_auc = compute_auc(train_true, knn_train_predictions)
print('')
print(f'KNN Accuracy: {knn_acc * 100} %')
print(f'KNN AUC: {knn_auc * 100} %')
print(f'KNN Train Accuracy: {knn_train_acc * 100} %')
print(f'KNN Train AUC: {knn_train_auc * 100} %')
```
